#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

```python
ax1 = fitbit.pivot_table(index=["Months_encoded"], values = ['Steps'], aggfunc={'Steps': np.median})
ax1 = ax1.reset_index()
x1 = ax1["Months_encoded"]
y1 = ax1["Steps"]
ax2 = fitbit.pivot_table(index=["Work_or_Weekend"], values = ['Steps'], aggfunc={'Steps': np.median})
ax2 = ax2.reset_index()
x2 = ax2["Work_or_Weekend"]
y2 = ax2["Steps"]

fig, [ax1, ax2] = plt.subplots(1, 2)
ax1.plot(x1,y1)
ax1.set_xlabel('Months')      
ax1.set_ylabel('Steps')      
ax1.set_title('Median steps per month')
ax2.plot(x2,y2)
ax2.set_xlabel('Weekend vs. Weekday')
ax2.set_ylabel('Steps')
ax2.set_title('Median steps for weekend/weekday')

plt.show()
```
