```python
import numpy as np
import pandas as pd
```

```python
#Importing Data as a CSV file into Python 


data = pd.read_csv('C:/Users/denis/Desktop/Ironhack/ironhackLabs/module-1/Data-Cleaning/datacopy.csv')

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
data.head(10)
```

```python
#Understanding the shape of the data
data.shape
```

```python
# Getting an understanding of the columns in the data 
##Need to examine whether these columns are correctly classified, need to change Year to object
data.columns
data.dtypes
```

```python
#examining missing values 
null_cols = data.isnull().sum()
null_cols[null_cols > 0]
```

```python
#Filtering out missing data 
# I decided to only remove Unnamed 22 & 23 due to missing data

drop_cols = list(null_cols[null_cols > 5000].index)
data = data.drop(drop_cols, axis=1)
# data.head()
```

```python
# Change Year from int to object: 
data['Year'] = data['Year'].astype('object')
data['Year'].dtype
```

```python
# Cleaning Columns
# The names, pdf, href, href formula, additional case numbers and original number columns are redundant or messy, 
#so we will scrap it. 
data = data.drop(["Name","pdf", "href", "href formula","Case Number.1", "Case Number.2", "original order" ], axis=1)
data.head()
```

```python
# Cleaning the Sex column 
#First, there's an extra space in Sex_, so we'll remove that
data = data.rename(columns={"Sex ":"Sex"})
data.columns

# We want to examine the items in the Sex column
data['Sex'].value_counts()

#I want to rename entries "M " and "lli" and remove entries "." and "N"
data.Sex = data.Sex.replace({"M ": "M", "lli": "M"})

data = data[data['Sex']!='.']
data = data[data['Sex']!='N']
#The Sex column has now been cleaned 
data['Sex'].value_counts()
```

```python
data = data.rename(columns={"Fatal (Y/N)":"Fatal"})
data.columns
```

```python
# cleaning Age
data['Age'].value_counts()
data = (data.drop(['Age'], axis=1)
         .join(data['Age'].apply(pd.to_numeric, errors='coerce')))
data.head(10) 


# data["Age"]= data["Age"].astype(str)
# # data["Age new"]= data["Age"].str.replace("A-Za-z", "") 
# data["bool_series1"]= data["Age"].str.isdigit()
# data["bool_series2"]= data["Age new"].str.isdigit() 
# data[data['Age']].rstrip('s')
# replace needs a specific value to match 
# data["Age"]= data["Age"].replace
```

```python
# Cleaning Fatal 

# data = data.rename(columns={"Fatal (Y/N)":"Fatal"})
# data.columns
data.Fatal = data.Fatal.replace({" N": "N", "N ": "N", "n":"N", "UNKNOWN": np.nan, "F": np.nan,"#VALUE!": np.nan})
data['Fatal'].value_counts()
```

```python
# Cleaning Type
data['Type'].value_counts()
data.Type = data.Type.replace({"Boating": "Boat"})
data['Type'].value_counts()
```
